# Merging Datasets
Use Pandas Merges to create a combined dataset from `clean_08.csv` and `clean_18.csv`.

```python
import pandas as pd
```

```python
# load datasets

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

### Create combined dataset

```python
# rename 2008 columns
df_08.rename(columns=lambda x: x[:] + '_2008', inplace=True)
```

```python
# rename 2018 columns
df_18.rename(columns=lambda x: x[:] + '_2018', inplace=True)
```

```python
# view to check names
df_08.head()
```

```python
# view to check names
df_18.head()
```

```python
# merge datasets
df_combined = df_08.merge(df_18, left_on='model_2008', right_on='model_2018', how='inner')
```

```python
# view to check merge
df_combined.head(1)
```

Save the combined dataset

```python
df_combined.to_csv('combined_dataset.csv', index=False)
```
